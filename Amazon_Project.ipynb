```python
#Type Conversion: release_year and runtime
#Ensure columns have the correct data type for analysis
df2['release_year'] = df2['release_year'].astype('Int64')
df2['runtime'] = df2['runtime'].astype('Float64')
print("After type conversion (release_year and runtime):")
display(df2.dtypes)

# Fill missing age certifications
df2['age_certification'].fillna('Unknown', inplace=True)
df2['age_certification'].fillna('Unknown', inplace=True)
print("After filling missing age certifications:")
display(df2['age_certification'].value_counts())

#Optional: confirm final cleaned datasets
print("Final Titles Dataset:")
display(df2.head())
print("Final Credits Dataset:")
display(df1.head())
```

```python
#Merging Titles and Credits Datasets--
# Merge titles(df2) with credits(df1) on 'id'
merged_df = pd.merge(
df2,df1,on='id',how='left')
print("merged completed", merged_df.shape)

#overview of merged dataset
print("\nColumns in merged dataset:")
print(merged_df.columns)

print("\nGeneral Info:")
merged_df.info()

print("Sample Rows:")
display(merged_df.head())

#post-merge cleaning
#to drop duplicates after merging
merged_df.drop_duplicates(subset=['id', 'person_id', 'role', 'name'], inplace=True)
print("Duplicates removed. New Shape:", merged_df.shape)

#handling missing values
print("misssing values")
merged_df.isnull().sum()

#fill missing values
merged_df['role'].fillna("Unknown", inplace=True)
merged_df['character'].fillna("N/A", inplace=True)
merged_df['name'].fillna("Unknown", inplace=True)

#text formatting
merged_df['role'] = merged_df['role'].str.title()
merged_df['name'] = merged_df['name'].str.title()
merged_df['character'] = merged_df['character'].str.title()

#missing patterns insights
print("MISSING DATA PATTERNS INSIGHTS ---")
# Missing character names
missing_characters = merged_df[merged_df['character'] == "N/A"].shape[0]
print(f"Missing character names (mostly crew members): {missing_characters}")

# Missing age certifications
missing_age = merged_df[merged_df['age_certification'] == "Unknown"].shape[0]
print(f"Missing or unknown age certifications: {missing_age}")

# Missing IMDb scores
missing_imdb_score = merged_df['imdb_score'].isnull().sum()
print(f"Missing IMDb scores: {missing_imdb_score}")

# Missing TMDB scores
missing_tmdb_score = merged_df['tmdb_score'].isnull().sum()
print(f"Missing TMDB scores: {missing_tmdb_score}")

#Business relevant insights
# Actor/Director Influence on Ratings (Top 10)
actor_rating = (merged_df.groupby('name')['imdb_score'].mean().sort_values(ascending=False)
                .head(10))
print("Top 10 Highest Rated Actors/Directors:")
print(actor_rating)

# Country-wise Rating Analysis
df_exploded_country = merged_df.explode('production_countries')
country_ratings = (df_exploded_country.groupby('production_countries')['imdb_score'].mean()
                   .sort_values(ascending=False).head(10))
print("Top Countries with Highest Rated Content:")
print(country_ratings)

#Genre-wise Actor Patterns
df_exploded_genres = merged_df.explode('genres')
top_genre_actors = (df_exploded_genres.groupby('genres')['name'].agg(lambda x: x.value_counts().index[0]).head(10))
print("Most Frequently Appearing Actors in Each Genre:")
print(top_genre_actors)

#Popular Cast vs TMDB Popularity
popularity_by_actor = (merged_df.groupby('name')['tmdb_popularity'].mean().sort_values(ascending=False).head(10))
print("Top 10 Actors with Highest TMDB Popularity Influence:")
print(popularity_by_actor)

```
